```python
#OPERACIONES DE REDUCCIÓN (sum, min, max...)

total_inversion_marketing=marketing_data["Inversión en marketing (USD)"].sum()
print(total_inversion_marketing)
```

```python
#INSERCIÓN DE UNA NUEVA COLUMNA EN EL DATASET

marketing_data["Utilidad"]=marketing_data['Ventas generadas (USD)']-marketing_data["Inversión en marketing (USD)"]
marketing_data
```

```python
#UNIÓN DE DATASETS
data_ejecutivo=pd.read_excel("/content/sample_data/Taller2_ETL.xlsx",sheet_name='Ejecutivos')
data_ejecuciones=pd.read_excel("/content/sample_data/Taller2_ETL.xlsx",sheet_name='BaseEjecuciones')
```

```python
data_ejecutivo
```

```python
data_ejecuciones
```

```python
#MERGE DE DATOS
ejecutivos_ejecuciones=pd.merge(data_ejecutivo,data_ejecuciones,how='inner',on="CEDULA")
ejecutivos_ejecuciones
```

```python
#TABLAS DINÁMICAS
#CALCULAR EL TOTAL DE TARJETAS DE CRÉDITO POR ZONA

Tabla_dinamica=pd.pivot_table(ejecutivos_ejecuciones,
               index='ZONA',    #FILAS DE LA TABLA DINÁMICA
               values='N° Tarjetas De Crédito',    #VALOR A TRAER
               aggfunc=np.sum)
Tabla_dinamica
```

```python
#ELIMINAR UNA COLUMNA EN EL DATASET
marketing_data=marketing_data.drop('Utilidad',axis="columns")
marketing_data
```

```python
#INSTALACIÓN DE LIBRERIAS
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#CARGAR UN DATA FRAME DENTRO DE UNA VARIABLE A TRAVÉS DE LA LIBRERIA PANDAS
comparacion_ventas_exp=pd.read_csv("/content/sample_data/comparacion_ventas_exp.csv")
composicion_mercado_exp=pd.read_csv("/content/sample_data/composicion_mercado_exp.csv")
correlacion_marketing_exp=pd.read_csv("/content/sample_data/correlacion_marketing_exp.csv")
distribucion_tiempos_exp=pd.read_csv("/content/sample_data/distribucion_tiempos_exp.csv")
```

**GRÁFICA DE CORRELACIÓN**

```python
correlacion_marketing_exp.head(5)
```

```python
#GRÁFICA DE CORRELACIÓN DE LA INVERSIÓN VS VENTAS GENERADAS

#df=sns.load_dataset("correlacion_marketing_exp")
sns.scatterplot(x=correlacion_marketing_exp["Inversión en marketing (USD)"],y=correlacion_marketing_exp["Ventas generadas (USD)"])
```

```python
#GRÁFICA DE CORRELACIÓN DE LA INVERSIÓN VS VENTAS GENERADAS
#ASIGNAR COLOR A OS PUNTOS
sns.scatterplot(x=correlacion_marketing_exp["Inversión en marketing (USD)"],y=correlacion_marketing_exp["Ventas generadas (USD)"],color="red")
```

```python
#GRÁFICA DE CORRELACIÓN DE LA INVERSIÓN VS VENTAS GENERADAS

sns.scatterplot(x=correlacion_marketing_exp["Inversión en marketing (USD)"],
                y=correlacion_marketing_exp["Ventas generadas (USD)"],
                hue=correlacion_marketing_exp["Campaña ID"],
                legend=True)
```

**GRÁFICA HEATMAP**

```python
sns.heatmap(correlacion_marketing_exp[['Campaña ID','Inversión en marketing (USD)','Ventas generadas (USD)']])
```
